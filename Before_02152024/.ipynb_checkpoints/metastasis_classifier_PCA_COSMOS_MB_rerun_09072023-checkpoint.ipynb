```python
# read metadata (COSMOS meta data sent by MY)
df_cosmos_meta = pd.read_csv("/ghess/pharma/projects/sirius_pharma/mrd_ist/cosmos/deep_dive/COS_01_Combined Report (Batch1,2&Enrollment&cfDNA retains)_TR working copy.csv", header=0)
df_cosmos_meta_simple = df_cosmos_meta.loc[:, ["GHSampleID", "Patient_ID", "Sample Timing to Treatment Milestones", "Visit_name", "Tumor_fraction", "Relapse site Site"]]
df_cosmos_meta_simple.columns = ["GHSampleID", "Patient_ID", "Timing_to_treatment", "Visit_name", "TF_S3_model", "met_site"]
print("Total sample size of the COSMOS cohort is " + str(df_cosmos_meta_simple.shape[0]) + "\n") 
print("Total unique sample size of the COSMOS cohort is " + str(df_cosmos_meta_simple.GHSampleID.nunique()) + "\n") 
print("Total unique patient size of the COSMOS cohort is " + str(df_cosmos_meta_simple.Patient_ID.nunique()) + "\n") 

#drop duplications (duplication is due to the one methyl TF corresponding with many somatic variants call in TB report)
df_cosmos_meta_simple = df_cosmos_meta_simple.drop_duplicates(subset='GHSampleID', keep='first')
print("Total unique sample size of the COSMOS cohort after deduplication is " + str(df_cosmos_meta_simple.GHSampleID.nunique()) + "\n") 

#filter metastastic cancer samples whose met site is either lung only, liver only or primary (no metastasis)
df_cosmos_meta_simple['met_site'].fillna("Primary", inplace=True) #name N/A as primary 
df_cosmos_meta_simple = df_cosmos_meta_simple[(df_cosmos_meta_simple["met_site"] == "Lung Only") | (df_cosmos_meta_simple["met_site"] == "Liver only") | (df_cosmos_meta_simple["met_site"] == "Primary")]
print(df_cosmos_meta_simple.met_site.value_counts())


#add TF by using MB model. 
#notes: the TF(after MB model binary call) was called "new_force_TF". Because COSMOS cohort used S3 model binary call. Now switched to MB model. 
#Even though the quantification model for TF prediction is the same of those two models, the binary call results might be different. So to make sure all the COSMOS MBresun has TF score, Sai created the "new_force_TF".
df_cosmos_MB_rerun = pd.read_table("/ghess/omni_v2/users/schen/mafband_on_s3/COSMOS_2031_samples_3_batches.with_metrics_and_mafband.20230816.tsv") 
df_cosmos_meta_simple = df_cosmos_meta_simple.merge(df_cosmos_MB_rerun[['GHSampleID', 'new_force_TF']], on='GHSampleID', how='left')
df_cosmos_meta_simple.rename(columns={'new_force_TF': 'TF_MB_model'}, inplace=True)

# #TF filter based on S3 model and corresponding TF call when S3 model is pos
# df_cosmos_meta_TF_filtered = df_cosmos_meta_simple[df_cosmos_meta_simple["TF_S3_model"] > 0.01]

#TF filter based on MB model and corresponding TF call when MB model is pos
df_cosmos_meta_TF_filtered = df_cosmos_meta_simple[df_cosmos_meta_simple["TF_MB_model"] > 0.01].reset_index(drop = True)
print("Total sample size of the COSMOS cohort whose TF > 0.01 and has metastatsis on primary, or lung only or liver only is " + str(df_cosmos_meta_TF_filtered.shape[0]) + "\n") 
print(str(df_cosmos_meta_TF_filtered.met_site.value_counts()) + "\n")
df_cosmos_meta_TF_filtered.head()

#TF match did not do becuase the sample size it too small
# df_meta_prim = df_cosmos_meta_TF_filtered[df_cosmos_meta_TF_filtered["met_site"] == "Primary"]
# df_meta_metastasis = df_cosmos_meta_TF_filtered[~(df_cosmos_meta_TF_filtered["met_site"] == "Primary")]

# print(df_meta_prim["TF_MB_model"].describe())
# print(df_meta_metastasis["TF_MB_model"].describe())
```

## Metastasis vs Primary data exploration (PCA and heatmap)

```python
#generate the PCA plot input data by aggregating COSMOS MBrerun region_ID and their norm counts of candidate samples
def get_mb_regionID_norm_counts(df_cosmos_MBrerun):
    mb_caller_result_folder = "/ghess/omni_v2/users/schen/mafband_on_s3/mafband_shield_bip"
    
    #an empty list to store individual DataFrames
    dfs = []
    
    for index, crc_row in df_cosmos_MBrerun.iterrows():
        run_sample_id = crc_row["GHSampleID"]
        fc_id = crc_row["fc_id"]
        msre_mr_file = f"{mb_caller_result_folder}/{fc_id}/{run_sample_id}/{run_sample_id}.msre_caller_mr_features.hdr.tsv.gz"
        # print(msre_mr_file)
        
        df = pd.read_table(msre_mr_file) 
        df["norm_count"] = df["hyper_residual_count"] / df["hyper_residual_count_pos_ctrl"]
        df = df.sort_values("region_id")
        run_sample_id_value = df["run_sample_id"].iloc[0]
        df.index = df.region_id
        df = df[["norm_count"]] 
        df = df.transpose()

        df.index = [run_sample_id_value]
        df = df.reset_index()
        df.rename(columns={'index': 'run_sample_id'}, inplace=True)
        df.columns.name = None
        
        #append the DataFrame to the list
        dfs.append(df)
        
    #concatenate all DataFrames in the list to create the final DataFrame
    df_final = pd.concat(dfs, ignore_index=True)

    return df_final


df_cosmos_MB_rerun = pd.read_table("/ghess/omni_v2/users/schen/mafband_on_s3/COSMOS_2031_samples_3_batches.with_metrics_and_mafband.20230816.tsv") 

#generate input data of all samples for PCA
metastasis_patient_list_all = df_cosmos_meta_simple.GHSampleID.tolist()
df_cosmos_MB_rerun_metastatsis_all = df_cosmos_MB_rerun[df_cosmos_MB_rerun.GHSampleID.isin(metastasis_patient_list_all)].reset_index(drop = True)
df_metastasis_MB_regions_all = get_mb_regionID_norm_counts(df_cosmos_MB_rerun_metastatsis_all)
print("Total sample size of the COSMOS cohort with MB rerun data is " + str(df_metastasis_MB_regions_all.shape[0]) + "\n") 

#generate input data of MB TF>0.01 samples for PCA
metastasis_patient_list = df_cosmos_meta_TF_filtered.GHSampleID.tolist()
df_cosmos_MB_rerun_metastatsis = df_cosmos_MB_rerun[df_cosmos_MB_rerun.GHSampleID.isin(metastasis_patient_list)].reset_index(drop = True)
df_metastasis_MB_regions_TF_filtered = get_mb_regionID_norm_counts(df_cosmos_MB_rerun_metastatsis)
print("Total sample size of the COSMOS cohort whose TF > 0.01 with MB rerun data is " + str(df_metastasis_MB_regions_TF_filtered.shape[0]) + "\n") 

#generate input data of baseline samples for PCA
metastasis_patient_list_base = df_cosmos_meta_base.GHSampleID.tolist()
df_cosmos_MB_rerun_metastatsis_base = df_cosmos_MB_rerun[df_cosmos_MB_rerun.GHSampleID.isin(metastasis_patient_list_base)].reset_index(drop = True)
df_metastasis_MB_regions_base = get_mb_regionID_norm_counts(df_cosmos_MB_rerun_metastatsis_base)
print("Total sample size of the COSMOS cohort at baseline with MB rerun data is " + str(df_metastasis_MB_regions_base.shape[0]) + "\n") 

#generate input data of landmark samples for PCA
metastasis_patient_list_landmark = df_cosmos_meta_landmark.GHSampleID.tolist()
df_cosmos_MB_rerun_metastatsis_landmark = df_cosmos_MB_rerun[df_cosmos_MB_rerun.GHSampleID.isin(metastasis_patient_list_landmark)].reset_index(drop = True)
df_metastasis_MB_regions_landmark = get_mb_regionID_norm_counts(df_cosmos_MB_rerun_metastatsis_landmark)
print("Total sample size of the COSMOS cohort at landmark with MB rerun data is " + str(df_metastasis_MB_regions_landmark.shape[0]) + "\n") 
df_metastasis_MB_regions_landmark.head()
```

```python
import warnings
warnings.simplefilter("ignore")
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
from sklearn.feature_selection import f_classif, chi2
from sklearn.feature_selection import SelectKBest
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn import model_selection

# # if there's a module named plotting_functions in the /ghds/mrd/analyses/notebooks/jkurata/gh-jkurata/jk_helper/ directory. Only needed when using make_scatter() method
# sys.path.append("/ghds/mrd/analyses/notebooks/jkurata/gh-jkurata/jk_helper/")
# from plotting_functions import *


#generate PCA plots of raw input data
def pca_plots(df_x, df_y):
    '''
    #df_x is the df with all the region_id and its corresponding norm_count plus one "run_sample_id" column
    #df_y is the df with metadata of the samples of interests (as long as it has column 'run_sample_id' and 'met_site')
    '''
    
    #PCA input data preparation
    x= df_x.drop(['run_sample_id'], axis=1)
    df_PCA_plot = df_x.copy() #It's important to do a deep copy to avoid edit the original df
    
    df_y_plot = df_y.copy() 
    df_y_plot.rename(columns={'GHSampleID': 'run_sample_id'}, inplace=True)
    df_PCA_plot = df_PCA_plot.merge(df_y_plot[['run_sample_id', 'met_site', 'TF_MB_model']], on='run_sample_id', how='left')
    # print(df_PCA_plot.head())
    pca_label = df_PCA_plot[['met_site']].reset_index(drop = True)
    pca_label = df_PCA_plot['met_site']
    y = pca_label

    #PCA training 
    pca = PCA(n_components=10)
    principalComponents = pca.fit_transform(np.log10(x + 0.0000001))
    
    #plot the variances of PCA features method1
    # features = range(pca.n_components_)
    # plt.bar(features, pca.explained_variance_ratio_) #use the pca.explained_variance_ to get the actual variance number rather tahn ratio
    # plt.xticks(features)
    # plt.ylabel('variance explained ratio')
    # plt.xlabel('PCA features')
    # plt.title("Variance explained within PCA features")
    # plt.show()
    # print('Explained variance ratio is ', pca.explained_variance_ratio_)
    # print("The cummulative variance sum is " + str(np.cumsum(pca.explained_variance_ratio_)))
    
    #plot the variances of PCA features method1
    print("The cummulative variance sum is " + str(np.cumsum(pca.explained_variance_ratio_)) + "\n")
    plt.plot(np.cumsum(pca.explained_variance_ratio_))
    plt.ylabel('Cumulative Explained Variance', fontsize=12)
    plt.xlabel('Number of components', fontsize=12)
    # plt.title('Plot showing number of components and the extent of variance accounted by them')
    plt.show()

    #PCA plot from Che-yu's code
    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['principal component 1', 'principal component 2',
                             'principal component 3', 'principal component 4',
                             'principal component 5', 'principal component 6',
                             'principal component 7', 'principal component 8',
                             'principal component 9', 'principal component 10'])

    finalDf = pd.concat([principalDf, df_PCA_plot[["run_sample_id", "met_site", 'TF_MB_model']]], axis = 1)
    per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
    labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
    print(per_var)
    print(labels)
    plt.bar(x=range(1, len(per_var)+1), height=per_var, tick_label=labels)
    plt.xlabel("Pricipal Component")
    plt.ylabel("% of explained variance")
    plt.show()

    sns.scatterplot(x="principal component 3", y = "principal component 2", hue="met_site", hue_order = ["Liver only", "Lung Only", "Primary"], palette = ["b", "r", "g"], data=finalDf)
    # sns.scatterplot(x="TF_MB_model", y = "principal component 1", hue="met_site", hue_order = ["Liver only", "Lung Only", "Primary"], palette = ["b", "r", "g"], data=finalDf)
    # plt.xlim([0, 0.004])
    #alternative plot method using Reghu's package
    # make_scatter(finalDf1, "principal component 1", "principal component 2", "", hue = "met_site", hue_order = ["Liver Only", "Lung Only", "Primary"], palette = ["b", "r", "g"])
 

# pca_plots(df_metastasis_MB_regions_all, df_cosmos_meta_simple)
# pca_plots(df_metastasis_MB_regions_TF_filtered, df_cosmos_meta_simple)
# pca_plots(df_metastasis_MB_regions_all_TF_removed, df_cosmos_meta_simple)
pca_plots(df_metastasis_MB_regions_landmark, df_cosmos_meta_simple)
# pca_plots(df_metastasis_MB_regions_base, df_cosmos_meta_simple) # running error due to small simple size
```

## PCA feature selection

```python
import warnings
warnings.simplefilter("ignore")
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
from sklearn.feature_selection import f_classif, chi2
from sklearn.feature_selection import SelectKBest
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler, Normalizer

from sklearn import model_selection
    
#build PCA model using f_classif and visualize the training data performance
def pca_model(df_x, df_y):

    #df_x is the df with all the region_id and its corresponding norm_count plus one "run_sample_id" column
    x = df_x.drop("run_sample_id", axis=1)

    #df_y is the df with metadata of the samples of interests (as long as it has column 'run_sample_id' and 'met_site')
    df_y_pca = df_y.copy() 
    df_y_pca.rename(columns={'GHSampleID': 'run_sample_id'}, inplace=True)
    df_y_pca = df_x.merge(df_y_pca[['run_sample_id', 'met_site']], on='run_sample_id', how='left')
    y = df_y_pca[['met_site']].reset_index(drop = True)
    y = y['met_site'] #convert to series
    
    # Reserve 20% for testing
    X_train, X_test, y_train, y_test = model_selection.train_test_split(df_x, df_y_pca, test_size=0.2,random_state=10)
    print('training data has %d observation with %d features'% X_train.shape)
    print('test data has %d observation with %d features'% X_test.shape)
    
    #data normalization
    x_scaled = x
    data_normalizer = Normalizer(norm='l2')
    x_scaled_norm = data_normalizer.fit_transform(x_scaled)
    df_x_scaled_norm = pd.DataFrame(x_scaled_norm, columns=x.columns)
    df_x_scaled_norm = df_x_scaled_norm.set_index(x.index)

    #model training by using f_classif
    select_k_best_fclassif_400 = pd.DataFrame(SelectKBest(f_classif, k=400).fit_transform(df_x_scaled_norm, y))
    print(select_k_best_fclassif_400.shape)
    select_k_best_fclassif_400_feature_index = SelectKBest(f_classif, k=400).fit(df_x_scaled_norm, y).get_support(indices=True)
    select_k_best_fclassif_400_feature_regions = df_x_scaled_norm.iloc[:, select_k_best_fclassif_400_feature_index].columns.tolist()
    len(select_k_best_fclassif_400_feature_regions)
    
    #save the list "select_k_best_fclassif_400_feature_regions" as tsv
    with open("/home/hazhang/projects/metastasis_classifier_PCA_090523/select_k_best_fclassif_400_feature_regions.tsv", "w") as output:
        for item in select_k_best_fclassif_400_feature_regions:
            output.write(item + '\n')
        
    #generate some pca plots
    pca = PCA(n_components=10)
    principalComponents = pca.fit_transform(np.log10(select_k_best_fclassif_400 + 0.0000001))
    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['principal component 1', 'principal component 2',
                             'principal component 3', 'principal component 4',
                             'principal component 5', 'principal component 6',
                             'principal component 7', 'principal component 8',
                             'principal component 9', 'principal component 10'])
    finalDf1 = pd.concat([principalDf, y], axis = 1)
    
    #save the training model by using pickle
    pickle.dump(pca, open(f"/home/hazhang/projects/metastasis_classifier_PCA_090523/fclassif_pca.pkl","wb"))
    per_var = np.round(pca.explained_variance_ratio_*100, decimals=1)
    labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
    print(per_var)
    print(labels)
    plt.bar(x=range(1, len(per_var)+1), height=per_var, tick_label=labels)
    plt.xlabel("Pricipal Component")
    plt.ylabel("% of explained variance")
    plt.show()
    sns.scatterplot(x="principal component 2", y = "principal component 3", hue="met_site", hue_order = ["Liver only", "Lung Only", "Primary"], palette = ["b", "r", "g"], data=finalDf1)
    #alternative plot method using Reghu's package
    # make_scatter(finalDf1, "principal component 1", "principal component 2", "", hue = "met_site", hue_order = ["Liver Only", "Lung Only", "Primary"], palette = ["b", "r", "g"])
    plt.show()
    
    # return X_test.loc[:, "run_sample_id"].reset_index(drop = True), y_test.reset_index(drop = True)
    return y_test.reset_index(drop = True)
    
    
# pca_model(df_metastasis_MB_regions_TF_filtered, df_cosmos_meta_simple)
# y_test = pca_model(df_metastasis_MB_regions_all, df_cosmos_meta_simple)
y_test = pca_model(df_metastasis_MB_regions_all_TF_removed, df_cosmos_meta_simple)
```

```python
#load model and 400 best regions with that model
pca_reload_fclassif = pickle.load(open(f"/home/hazhang/projects/metastasis_classifier_PCA_090523/fclassif_pca.pkl",'rb'))
select_k_best_fclassif_400_feature_regions = pd.read_table("/home/hazhang/projects/metastasis_classifier_PCA_090523/select_k_best_fclassif_400_feature_regions.tsv", header=None)
select_k_best_fclassif_400_feature_regions = select_k_best_fclassif_400_feature_regions[0].tolist()

def pca_classifier_PC1_PC2_mb(df_msre_caller_mr_features):
    df = pd.read_table(df_msre_caller_mr_features)
    df["norm_count"] = df["hyper_residual_count"] / df["hyper_residual_count_pos_ctrl"]
    df = df.sort_values("region_id")
    df.index = df.region_id
    df = df[["norm_count"]] 
    df = df.transpose()
    
    data_normalizer = Normalizer(norm='l2')
    df_scaled_norm = data_normalizer.fit_transform(df)
    df_scaled_norm = pd.DataFrame(df_scaled_norm, columns=df.columns)
    df_scaled_norm = df_scaled_norm.set_index(df.index)

    result_new = pca_reload_fclassif.transform(np.log10(df_scaled_norm[select_k_best_fclassif_400_feature_regions] + 0.0000001))
    # print(result_new)
    return result_new[0][0], result_new[0][1]

pca_classifier_PC1_PC2_mb("/ghess/omni_v2/users/schen/mafband_on_s3/mafband_shield_bip/230324_A01902_0106_AHVHYFDSX5_PS3R212_AP2.Sirius-1.0.0-RLS/A025425210/A025425210.msre_caller_mr_features.hdr.tsv.gz")    
```
